```python
files = ['LDA_SSI_results.csv','SVML_SSI_results.csv','SVMR_SSI_results.csv',
         'MLP1_SSI_results.csv','MLP2_SSI_results.csv','MLP3_SSI_results.csv',
         'NN1_SSI_results.csv','NN2_SSI_results.csv','NN3_SSI_results.csv']
names = ['LDA','SVML','SVMR','MLP1','MLP2','MLP3','NNET1','NNET2','NNET3']

t1_flag = 0
if(t1_flag==1):
    for j in range(len(names)):
        clf = names[j]
        table_name = 'SSI_Results_Tables\\' + 'AA_' + clf + '_SSI_Table.txt'
        with open(table_name,'a') as f:
            f.write(r'\begin{table}[]'),f.write('\n')
            title = '\caption{'+clf+' Single Subject Independent Results.'+'}'
            f.write(title),f.write('\n')
            f.write(r'\centering'),f.write('\n')
            f.write(r'\setlength{\tabcolsep}{5pt}'),f.write('\n')
            f.write(r'\renewcommand{\arraystretch}{1}'),f.write('\n')
            f.write(r'\begin{tabular}{c|cccc}'),f.write('\n')
            f.write(r'\hline'),f.write('\n')
            f.write(r'\hline'),f.write('\n')
            f.write(r'\textbf{Subject} & \textbf{Mean} & \textbf{Std} & \textbf{Min} & \textbf{Max} \\') ,f.write('\n')
            f.write(r'\hline'),f.write('\n')
            f.write(r'\hline'),f.write('\n')
            
            ############################## TABLE CONTENTS ##############################
            path = 'SSI\\' + files[j]
            with open(path,'r') as content:
                df = pd.read_csv(content)
                for i in range(len(df)):
                    row = df.iloc[i] 
                    if i < 10 and i <30:
                        s = 'S0' + str(i)
                        f.write(s +
                                ' & ' + str(row[0])+'\%' +
                                ' & ' + r'\textpm'+str(row[1]) +
                                ' & ' + str(row[2])+'\%' +
                                ' & ' + str(row[3])+'\%' +
                                r' \\' + '\n')
                        f.write(r'\hline'),f.write('\n')
                    elif i >= 10 and i <30:  
                        s = 'S'+str(i)
                        f.write(s +
                                ' & ' + str(row[0])+'\%' +
                                ' & ' + r'\textpm'+str(row[1]) +
                                ' & ' + str(row[2])+'\%' +
                                ' & ' + str(row[3])+'\%' +
                                r' \\' + '\n')
                        f.write(r'\hline'),f.write('\n')
                    if i == 30:
                        s = 'All'
                        f.write(r'\textbf{'+s+'}' +
                                '&' + r'\textbf{'+str(row[0])+'\%}' +
                                '&' + r'\textbf{'+r'\textpm'+str(row[1])+'}' + 
                                '&' + r'\textbf{'+str(row[2])+'\%}' +
                                '&' + r'\textbf{'+str(row[3])+'\%}' +
                                r' \\' + '\n')
                        f.write(r'\hline'),f.write('\n')
            ############################################################################
            
            f.write(r'\hline'),f.write('\n')
            f.write(r'\end{tabular}'),f.write('\n')
            f.write(r'\end{table}'),f.write('\n')
        
```
