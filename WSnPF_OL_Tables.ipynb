```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import csv
```

```python
files = ['MLP1_PartialFit.csv','MLP2_PartialFit.csv','MLP3_PartialFit.csv']
names = ['MLP1','MLP2','MLP3']

create_table = 0

if create_table == 1:
    for j in range(len(names)):
        clf = names[j]
        table_name = 'PFT_Results_Tables\\' + clf + '_PF_Table.txt'
        with open(table_name,'a') as f:
            f.write(r'\begin{table}[]'),f.write('\n')
            title = '\caption{'+clf+' Partial Fit Results'+'}'
            f.write(title),f.write('\n')
            f.write(r'\centering'),f.write('\n')
            f.write(r'\setlength{\tabcolsep}{5pt}'),f.write('\n')
            f.write(r'\renewcommand{\arraystretch}{1}'),f.write('\n')
            f.write(r'\begin{tabular}{c|cccccccccc}'),f.write('\n')
            f.write(r'\hline'),f.write('\n')
            f.write(r'\textbf{Subject} & \textbf{0\%} & \textbf{10\%} & \textbf{20\%} & \textbf{30\%} & \textbf{40\%} & \textbf{50\%} & \textbf{60\%} & \textbf{70\%} & \textbf{80\%} & \textbf{90\%} \\') ,f.write('\n')
            f.write(r'\hline'),f.write('\n')
            
            ############################## TABLE CONTENTS ##############################
            path = 'PFT\\' + clf + '_PartialFit.csv'
            with open(path,'r') as content:
                df = pd.read_csv(content)
                for i in range(len(df)):
                    
                    row = df.iloc[i]
                    row = np.array(row)
                    row = np.round(row,1)
                    
                    if i < 10 and i <30:
                        s = 'S0' + str(i)
                        f.write(s+
                                ' & ' + str(row[0])+
                                ' & ' + str(row[1])+
                                ' & ' + str(row[2])+
                                ' & ' + str(row[3])+
                                ' & ' + str(row[4])+
                                ' & ' + str(row[5])+
                                ' & ' + str(row[6])+
                                ' & ' + str(row[7])+
                                ' & ' + str(row[8])+
                                ' & ' + str(row[8])+
                                r' \\' + '\n')
                        f.write(r'\hline'),f.write('\n')
                    elif i >= 10 and i <30:  
                        s = 'S'+str(i)
                        f.write(s+
                                ' & ' + str(row[0])+
                                ' & ' + str(row[1])+
                                ' & ' + str(row[2])+
                                ' & ' + str(row[3])+
                                ' & ' + str(row[4])+
                                ' & ' + str(row[5])+
                                ' & ' + str(row[6])+
                                ' & ' + str(row[7])+
                                ' & ' + str(row[8])+
                                ' & ' + str(row[8])+
                                r' \\' + '\n')
                        f.write(r'\hline'),f.write('\n')
                    if i == 30:
                        s = 'All'
                        f.write(r'\textbf{'+s+'}' +
                                '&' + r'\textbf{'+str(row[0])+'\%}' +
                                '&' + r'\textbf{'+str(row[1])+'\%}' +
                                '&' + r'\textbf{'+str(row[2])+'\%}' +
                                '&' + r'\textbf{'+str(row[3])+'\%}' +
                                '&' + r'\textbf{'+str(row[4])+'\%}' +
                                '&' + r'\textbf{'+str(row[5])+'\%}' +
                                '&' + r'\textbf{'+str(row[6])+'\%}' +
                                '&' + r'\textbf{'+str(row[7])+'\%}' +
                                '&' + r'\textbf{'+str(row[8])+'\%}' +
                                '&' + r'\textbf{'+str(row[8])+'\%}' +
                                r' \\' + '\n')
                        f.write(r'\hline'),f.write('\n')
            ############################################################################
            
            f.write(r'\end{tabular}'),f.write('\n')
            f.write(r'\end{table}'),f.write('\n')
```
